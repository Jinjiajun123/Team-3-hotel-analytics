# 1. Data Preparation and Pipeline Setup

This notebook demonstrates how to build the SQLite database from the raw `review.json` data dump, and subsequently enrich the database using our Zero-Shot ML pipeline.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd()
if (PROJECT_ROOT / "src").exists():
    sys.path.insert(0, str(PROJECT_ROOT))
else:
    PROJECT_ROOT = PROJECT_ROOT.parent
    sys.path.insert(0, str(PROJECT_ROOT))

from src.utils import DB_PATH, SAMPLE_DB_PATH
from src.data_processing import build_database, build_sample_database
from src.ml_pipeline import enrich_reviews_with_ml
```

## 1.1 Build the Full Database

Parse the raw JSON, filter out invalid/old reviews, calculate initial statistics, and load them into a relational schema.

```python
# Build the primary database (this may take a few minutes if the JSON is large)
total_reviews = build_database(db_path=DB_PATH)
print(f"Total reviews in primary database: {total_reviews}")
```

## 1.2 Build the Sample Database

Generate a smaller subset (e.g. 5,000 reviews) for testing and rapid prototyping.

```python
# Build the smaller sample database
sample_reviews = build_sample_database(source_db=DB_PATH, sample_db=SAMPLE_DB_PATH, sample_size=10000)
print(f"Total reviews in sample database: {sample_reviews}")
```

## 1.3 Machine Learning Enrichment

Run the HuggingFace DistilBERT Zero-Shot classifier to tag semantic metadata (e.g., `is_luxury`, `is_budget`, `is_business`) onto the reviews. Since this is computationally expensive on CPU, we run it once here during the ETL phase so that our dashboards remain performant.

```python
# Enrich the sample database
enrich_reviews_with_ml(db_path=SAMPLE_DB_PATH)
```
